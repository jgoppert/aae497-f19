# Motion Capture Data Analysis

Here we plot the data recorded in lab using the pandas library.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/Big_Drone_Flight_Test_0906.csv', skiprows=(0, 1, 2, 3, 4, 5), usecols=(1, 6, 7, 8), index_col=0, dtype='float')
```

```python
data['X.1'].plot()
data['Y.1'].plot()
data['Z.1'].plot()
plt.grid()
plt.legend()
```

```python
plt.plot(data['X.1'], data['Y.1'])
plt.axis('equal')
plt.grid()
```
