# Plot of search methods

```python
import matplotlib.pyplot as plt

resolution_early = [1, 2, 3, 4, 5, 6, 7, 8]
resolution_late = [16, 64, 125, 250, 256, 500, 1000]
time_early = [29713, 50468, 51010, 29677, 12835, 30327, 29750, 11159]
time_late = [26501, 30201, 51430, 29706, 33338, 29963, 14363]

plt.scatter(resolution_early,time_early)
plt.xlabel('resolution (m)') 
plt.ylabel('time (ns)') 
plt.title('Time elapsed to find positions') 
plt.show()

plt.scatter(resolution_late,time_late)
plt.xlabel('resolution (m)') 
plt.ylabel('time (ns)') 
plt.title('Time elapsed to find positions') 
plt.show()
```
