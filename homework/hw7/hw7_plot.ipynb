```python
import numpy as np
from matplotlib import pyplot as plt

filename = 'plot.txt'
data = open(filename,'r')
lines = data.readlines

numbers = [[i for i in line.split()] for line in data.readlines()]

print(numbers)
time=[]
res=[]

for j in numbers:
    time.append(float(j[0]))
    res.append(float(j[1]))
```

```python
plt.plot(res, time)
plt.xlabel('resolution')
plt.ylabel('time (ns)')
plt.title('Result')
plt.show()
```
